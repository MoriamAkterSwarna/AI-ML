## **Starter Code**

```python
# You must run this code and use that data set for all the questions
import numpy as np
import pandas as pd

np.random.seed(42)

ids = np.arange(1, 11)
ages = np.random.randint(18, 60, 10)
salaries = np.random.randint(30000, 90000, 10)
departments = np.array(["HR", "IT", "Finance", "IT", "HR", "Sales", "Finance", "IT", "Sales", "HR"])

DF = pd.DataFrame({
"id": ids,
"age": ages,
"salary": salaries,
"dept": departments
})

DF.to_csv("employees.csv", index=False)
print("Sample Data Created and Saved as employees.csv")

```

# **Answer to the Q1**

```python
ages_array = np.array(ages)
salaries_array = np.array(salaries)

print("Ages Array:")
print(ages_array)
print("\n")
print("Salaries Array:")
print(salaries_array)

print("\n")
print("Ages: ")
print("dtype: ", ages_array.dtype, "\nndim: ", ages_array.ndim, "\nshape: ", ages_array.shape, "\nSize: ", ages_array.size )
print("\n")
print("Salaries: ")
print("dtype: ",salaries_array.dtype , "\nndim: ", salaries_array.ndim, "\nshape: ", salaries_array.shape, "\nSize: ", salaries_array.size)
```

## **Answer to the Q2**

```python
print("Highest Salary: " , salaries_array.max())
print("Lowest Salary: " , salaries_array.min())

print("\n")
print("Average Salary: " , salaries_array.mean())
print("Average Age: ", ages_array.mean())

```

## **Answer to the Q3**

```python
greater_30 = ages_array[ages_array > 30]
print("Ages greater than 30: ", greater_30)

print("Employees older than 30: ", greater_30.size)
```

## **Answer to the Q4**

```python
new_ages_arr = ages_array + 5
print("Previos Original Array: " , ages_array)
print("New Ages Array: ", new_ages_arr)
```

## **Answer to the Q5**

```python
total_salary = salaries_array.sum()
print("Total Salary: ", total_salary)

max_salary = salaries_array.max()
min_salary = salaries_array.min()

diff = max_salary - min_salary
print("Difference between Max and Min Salary: ", diff)
```
